```python
import pandas as pd
df=pd.read_csv('jobss.csv')
df.head()
```

```python
df.loc[df.sal<=2000, 'salary range']='<=2000'
df.loc[(df.sal>2000) & (df.sal<=5000), 'salary range']='2000-5000'
df.loc[df.sal>5000, 'salary range']='>5000'
df.head()
```

```python
new_cols=['salary 2000-5000', 'salary <2000', 'salary >=2000']
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()
encoder_df=pd.DataFrame(encoder.fit_transform(df[['salary range']]).toarray(), columns=new_cols)
final_df=df.join(encoder_df)
final_df.head()
```

```python
new_cols=['salary label']
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
encoder_df=pd.DataFrame(encoder.fit_transform(df[['salary range']]), columns=new_cols)
final_df=df.join(encoder_df)
final_df.head()
```
